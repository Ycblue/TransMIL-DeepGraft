```python
yaml_dir = '/homeStor1/ylan/data/sustainability/electricitymaps-contrib/config/zones/'
iso_a2_list = list(europe_shp.ISO_A2)


co2eq_parameters_path = '/homeStor1/ylan/Sustainability/susai/experiment-impact-tracker/experiment_impact_tracker/emissions/data/co2eq_parameters.json'
with open(co2eq_parameters_path, 'r') as stream:
    co2eq_parameters_dict = json.load(stream)

# print(co2eq_parameters_dict['fallbackZoneMixes'])

additional_region_info = {'IT': [], 'NO':[], 'ES': [], 'DK': [], 'CH':[0.499]}
for i in co2eq_parameters_dict['fallbackZoneMixes'].keys():
    # print(i)
    if 'renewableRatio' in co2eq_parameters_dict['fallbackZoneMixes'][i].keys():
        if 'IT' in i:
            # print(i)
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes'][i]['renewableRatio']
            additional_region_info['IT'].append(renewable_ratio)
        elif 'NO' in i:
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes'][i]['renewableRatio']
            additional_region_info['NO'].append(renewable_ratio)
        elif 'DK' in i:
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes'][i]['renewableRatio']
            additional_region_info['DK'].append(renewable_ratio)
        elif 'ES' in i:
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes'][i]['renewableRatio']
            additional_region_info['ES'].append(renewable_ratio)
            
# print(additional_region_info)
additional_region_info = {k: {'renewableRatio': sum(v)/len(v)} for k, v in additional_region_info.items()}
# print(additional_region_info)

modified_dict = co2eq_parameters_dict['fallbackZoneMixes']
modified_dict.update(additional_region_info) # add aggregated numbers into dictionary
# print(modified_dict)


no_statistics = []
no_renewable_ratio = []
data_available = {}
# print(len(iso_a2_list))
# print(iso_a2_list)
list_of_rr = []
for index, row in europe_shp.iterrows():
    # print(row['ISO_A2'])
    iso_a2 = row['ISO_A2']
# for iso_a2 in iso_a2_list:
    if iso_a2 in co2eq_parameters_dict['fallbackZoneMixes'].keys():
        # print(co2eq_parameters_dict['fallbackZoneMixes'][iso_a2].keys())
        if 'renewableRatio' in co2eq_parameters_dict['fallbackZoneMixes'][iso_a2].keys():
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes'][iso_a2]['renewableRatio']
            # print(renewable_ratio)
            data_available[iso_a2] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        else:
            # print(iso_a2, 'no renewableRatio')
            no_renewable_ratio.append(iso_a2)
            data_available[iso_a2] = 0
            list_of_rr.append(0)

    else:
        if row['SOV_A3'] ==  'GB1':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['GB']['renewableRatio']
            data_available['GB'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        # print(iso_a2, 'no statistics')
        elif row['SOV_A3'] == 'ITA':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['IT']['renewableRatio']
            data_available['IT'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'ESP':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['ES']['renewableRatio']
            data_available['ES'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'PRT':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['PT']['renewableRatio']
            data_available['PT'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'NL1':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['NL']['renewableRatio']
            data_available['NL'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'BEL':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['BE']['renewableRatio']
            data_available['BE'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'BIH':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['BA']['renewableRatio']
            data_available['BA'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'SRB':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['RS']['renewableRatio']
            data_available['RS'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        elif row['SOV_A3'] == 'FR1':
            renewable_ratio = co2eq_parameters_dict['fallbackZoneMixes']['FR']['renewableRatio']
            data_available['FR'] = renewable_ratio
            list_of_rr.append(renewable_ratio)
        else:
            no_statistics.append(iso_a2)
            data_available[iso_a2] = 0
            list_of_rr.append(0)

# # print(no_statistics)
# # print(no_renewable_ratio)
# print(data_available)
# print(len(list_of_rr))

europe_shp['renewableRatio'] = list_of_rr



    
```
